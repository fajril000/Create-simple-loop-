```python
import cv2
import matplotlib.pyplot as plt
from skimage.filters import roberts
import numpy as np

# Baca citra
gambar = cv2.imread('winter.jpg')

# Konversi citra ke grayscale jika belum grayscale
if len(gambar.shape) == 3:  # Jika gambar memiliki 3 channel (RGB)
    gambar_gray = cv2.cvtColor(gambar, cv2.COLOR_BGR2GRAY)
else:
    gambar_gray = gambar

# Deteksi tepi menggunakan Operator Roberts
tepi_roberts = np.abs(roberts(gambar_gray))

# Tampilkan citra asli dan hasil deteksi tepi
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(gambar, cv2.COLOR_BGR2RGB))
plt.title('Citra Asli')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(tepi_roberts, cmap='gray')
plt.title('Deteksi Tepi Roberts')
plt.axis('off')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Baca citra
originalImage = cv2.imread('winter.jpg', cv2.IMREAD_COLOR)

# Tentukan parameter untuk operator Canny
sigma = 1.9  # Parameter deviasi standar untuk Gaussian
thresholdLow = 0.05  # Batas bawah untuk non-maximum suppression
thresholdHigh = 0.15  # Batas atas untuk non-maximum suppression

# Aplikasikan Gaussian Blur untuk mengurangi noise
blurredImage = cv2.GaussianBlur(originalImage, (5, 5), sigma)

# Deteksi tepi dengan operator Canny
edgeImage = cv2.Canny(blurredImage, thresholdLow * 255, thresholdHigh * 255)

# Tampilkan citra asli dan citra hasil deteksi tepi
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(originalImage, cv2.COLOR_BGR2RGB))
plt.title('Citra Asli')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(edgeImage, cmap='gray')
plt.title('Deteksi Tepi dengan Canny')
plt.axis('off')

plt.show()
```
